```python
# Importing required libraries
import pandas as pd
import numpy as np
```

```python
# Load 1000 entries from our huge Dataset
df = pd.read_csv("2010-capitalbikeshare-tripdata.csv", nrows= 1000)
df.head()
```

# Convertion of Start Station addresses to the coordinates (latitude and longitude) 

It fetches the coordinates accurately, but it is too slow. So, we will fetch coordinates for just 1000 entries from our dataset. Ignore the warnings and service timeouts that you get while this model is running. At the end, you just have to make sure that the code is completely executed.

```python
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
import time
from geopy.exc import GeocoderTimedOut


location = []
def locationLocate():
    for i in range(0,len(df["Start station"])):
        try:
            locator = Nominatim(user_agent="streamlitdemo")
            geocode = RateLimiter(locator.geocode, min_delay_seconds=1)
            location.append(geocode(df['Start station'][i]))
        except:
            pass
locationLocate()
            
        

```

# Loading the coordinates into lists so as to add it to the existing data frame.

```python
lat = []
lon = []
for i in range(0,len(location)):
    if(location[i]):
        lat.append(location[i].latitude)
        lon.append(location[i].longitude)
    else:
        lat.append(np.nan)
        lon.append(np.nan)
print("latitude: ",len(lat))
print("longitude: ",lon)
```

```python
type(lat[426])
# lat[426]
```

```python
df['lat'] = lat # Adding latitude to the dataframe
```

```python
df['lon'] = lon # Adding longitude to the dataframe
```

```python
# Convert Nan string values to numpy nan values. Now, these Numpy nan values can be easily filtered out in the next step.
df = df.replace('Nan', np.nan)
type(df['lat'][426])
df['lat'][426]
```

# Dropping the rows containing nan values and export the data to a csv file.

```python
df.dropna().to_csv("Trip history with locations.csv") 
```
